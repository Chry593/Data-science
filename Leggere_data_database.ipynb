```python
import pandas as pd
import sqlite3
```

### **Connection**
Per prima cosa tocca stabilire la connession col database tramite **connect**

```python
conn = sqlite3.connect("/content/drive/MyDrive/Colab Notebooks/chinook.db")
```

Una volta creata la connessione dobbiamo creare un oggetto **Cursor**. Cursor ci aiuterà nell'esecuzione delle SQL queries

```python
cur = conn.cursor()
```

IL cursore creato ha il metodo **execute**, che  riceverà un parametro SQL per runnarlo nel database.
Il codice sotto prende le prime 5 righe dalla tabella employees:

```python
cur.execute("SELECT * FROM employees LIMIT 5")
```

Per assegnare il valore alla query bisogna runnare un altro comando, **fetchall**:

```python
result = cur.fetchall()
result
```

Ci restituirà una lista di tuple, dove ogni tupla corrisponde a una riga del database. Avere a che fare con i dati in questo modo è molto difficile, ma con pandas tutto diventa più facile:

```python
df = pd.DataFrame(result)
```

```python
df.head()
```

Prima di andare avanti è buona cosa chiudere la connessione con gli oggetti così da evitare problemi:

```python
cur.close()
conn.close()
```

### **Using pandas read_sql method**
Possiamo usare la funzione di pandas *read_sql* per leggere i risultati di una query SQL direttamente dento un dataFrame. Ha molti vantaggi come:
- non richiede di creare un cursore o di usare fetchall
- legge automaticamente i nome degli headers dalla tabella
- crea DataFrame, così possiamo esplorare velocemente i dati

```python
conn = sqlite3.connect("/content/drive/MyDrive/Colab Notebooks/chinook.db")
```

```python
df = pd.read_sql("Select * FROM employees;",conn) #primo parametro query, secondo oggetto conn
```

```python
df.head()
```

```python
df = pd.read_sql("Select * FROM employees;",conn,
                 index_col ="EmployeeId",
                 parse_dates = ["BirthDate","HireDate"]) #primo parametro query, secondo oggetto conn
df.head()
```

```python
df.info()
```

```python
df["ReportsTo"].isna().sum()
```

```python
df["ReportsTo"].mean()
```

```python
df["ReportsTo"] = df["ReportsTo"].fillna(df["ReportsTo"].mean())
```

```python
df
```

```python
df["ReportsTo"] > 1.75
```

```python
df["City"] = df["City"].astype("category")
df.info()
```

### **Using pandas real_sql_query method**
Possiamo ottenere lo stesso risultato usando **real_sql_query** :

```python
conn = sqlite3.connect("/content/drive/MyDrive/Colab Notebooks/chinook.db")
```

```python
df = pd.read_sql_query("SELECT * FROM  employees LIMIT 5",conn)
```

```python
df.head()
```

```python
df = pd.read_sql_query("SELECT * FROM  employees LIMIT 5",conn,
                       index_col = "EmployeeId",
                       parse_dates = ["BirthDate", "HireDate"])
df.head()
```

## **Using read_sql_table method**
**read_sql_table** è una funzione utile, ma funziona solo con SQLAlchemy, un PythonSQL toolkit e Object Realtional Mapper

```python
from sqlalchemy import create_engine
```

```python
engine = create_engine("sqlite:////content/drive/MyDrive/Colab Notebooks/chinook.db")
connection = engine.connect()
```

```python
df = pd.read_sql_table("employees", con = connection,
                       index_col = "EmployeeId",
                       parse_dates = ["BirthDate","HireDate"])
```

```python
df.head()
```

```python
connection.close()
```
