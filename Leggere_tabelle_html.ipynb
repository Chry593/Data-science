### **Esempio Difficile**
Possiamo anche usare il modulo **request** per avere il codice HTML di una pagina. Nella pagina che analizzeremo saranno presenti piÃ¹ di una tabella, noi vogliamo solo la tabella con le informazioni di ogni stagione.

```python
import requests
```

```python
html_url = "https://en.wikipedia.org/wiki/The_Simpson"
```

```python
r = requests.get(html_url)
wiki_tables = pd.read_html(r.text, header = 0)
```

```python
len(wiki_tables)
```

```python
simpsons = wiki_tables[2]
simpsons.head()
```

Veloce pulita: rimuovere header e settare Season come index :

```python
simpsons.drop([0,1], inplace=True)
```

```python
simpsons.set_index("Season", inplace=True)
```

Quale stagione ha avuto il numero minore di episodi?
